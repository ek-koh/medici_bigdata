```python
def bar_chart(feature):
   survived = train[train['Survived']==1][feature].value_counts()
   dead = train[train['Survived']==0][feature].value_counts()
   df = pd.DataFrame([survived,dead])
   df.index = ['Survived','Dead']
   df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
train['Age'].fillna(train.groupby('Pclass')['Age'].transform('median'), inplace=True)
test['Age'].fillna(train.groupby('Pclass')['Age'].transform('median'), inplace=True)
```

```python
train_test_data = [train, test]
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
    
train.isnull().sum()
```

```python
train.head(100)
```

```python
for dataset in train_test_data:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

#A-Za-z ?

train['Title'].value_counts()
```

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3, "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
for dataset in train_test_data:   dataset['Title'] = dataset['Title'].map(title_mapping)
```

```python
train.head(100)
```

```python
bar_chart('Title')
```

```python
for dataset in train_test_data:
    dataset.loc[dataset['Age']<=16, 'Age'] = 0
    dataset.loc[(dataset['Age']>16) & (dataset['Age']<=26), 'Age'] = 1
    dataset.loc[(dataset['Age']>26) & (dataset['Age']<=36), 'Age'] = 2
    dataset.loc[(dataset['Age']>36) & (dataset['Age']<=62), 'Age'] = 3
    dataset.loc[dataset['Age']>62, 'Age'] = 4
```

```python
bar_chart('Age')
```

```python
sex_mapping = {"male" : 0, "female" : 1}
for dataset in train_test_data:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
    
train.head(100)
```

```python
embarked_mapping = {"S" : 0, "C" : 1, "Q" : 2}
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].map(embarked_mapping)
    
train.head(100)
```

```python
train.drop('Cabin', axis=1, inplace=True)
test.drop('Cabin', axis=1, inplace=True)
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)
train.drop('Ticket', axis=1, inplace=True)
test.drop('Ticket', axis=1, inplace=True)
train.drop('Fare', axis=1, inplace=True)
test.drop('Fare', axis=1, inplace=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.naive_bayes import GaussianNB 
from sklearn.svm import SVC
import numpy as np

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
train_data = train.drop('Survived', axis = 1)
target = train['Survived']

test_data = test.drop('PassengerId', axis = 1).copy()
test_data.head()

train_data = train_data.drop('PassengerId', axis = 1).copy()
train_data.head()
```

```python
clf = KNeighborsClassifier(n_neighbors = 13)
#    Knn   
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
round(np.mean(score)*100, 2)
```

```python
clf = GaussianNB()
clf.fit(train_data, target)

prediction = clf.predict(test_data)
submission = pd.DataFrame({
    "PassengerId" : test["PassengerId"],
    "Survived" : prediction
})

submission.to_csv('submission.csv', index = False)
```
